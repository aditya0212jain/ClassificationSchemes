```python
import numpy as np
import scipy
from scipy.stats import norm
import math
import random
import operator
import csv
```

```python
class knn:
    """kNearest Neighbour classification implementation"""
    
    def __init__(self,k):
        self.k = k
        
    def euclideanDistance(self,data1,data2,n):
        distance = 0
        for i in range(0,n):
            distance+=np.square(int(data1[i])-int(data2[i]))
        return np.sqrt(distance)
    
    def manhattanDistance(self,data1,data2,n):
        distance = 0
        for i in range(n):
            distance+=abs(int(data1[i])-int(data2[i]))
        return distance
    
    def hammingDistance(self,data1,data2,n):
        distance = 0
        for i in range(n):
            if int(data1[i])==int(data2[i]):
                distance+=1
        return distance
    
    ##using euclidean distance for now
    def getNeighbours(self,dataset,data):
        distances = []
        length = len(data)
        for i in range(0,len(dataset)):
            ##change the below line for different distances
            dist = self.hammingDistance(data,dataset[i],length)
            distances.append((dataset[i],dist))
        distances.sort(key=operator.itemgetter(1))
        neighbours = []
        for i in range(self.k):
            neighbours.append(distances[i])
        return neighbours
    
    def getClass(self,neighbours):
        classOccurrence = {}
        for i in range(len(neighbours)):
            t = neighbours[i][0][-1]
            if t in classOccurrence:
                if (neighbours[i][1]!=0):
                    classOccurrence[t]+= 1.0/neighbours[i][1]
                else:
                    return t
            else:
                if (neighbours[i][1]!=0):
                    classOccurrence[t] = 1.0/neighbours[i][1]
                else:
                    return t
        sortedOccurrence = sorted(classOccurrence.items(),key=operator.itemgetter(1),reverse=True)
        return sortedOccurrence[0][0]
    
    def classification(self,dataset,data):
        nei = self.getNeighbours(dataset,data)
        ans = self.getClass(nei)
        return ans
```

```python
with open('../../railwayBookingList.csv','r') as csv_file:
    csv_reader = list(csv.reader(csv_file,delimiter=","))
#     csv_dicReader = csv.DictReader(csv_file)
    my_data = np.array(csv_reader)
    
    x0 = []
    x1 = []
    
    for i in range(1,len(my_data)):
        #print (my_data[i])
        a = my_data[i][1]
        my_data[i][1]= my_data[i][6]
        my_data[i][6]= a
        #exchange age category with class
        
        if (my_data[i][5]=='male'):
            my_data[i][5]=1
        else:
            my_data[i][5]=0
        
        if (my_data[i][4]=='FIRST_AC'):
            my_data[i][4]=1
        elif (my_data[i][4]=='NO_PREF'):
            my_data[i][4]=0
        elif (my_data[i][4]=='SECOND_AC'):
            my_data[i][4]=2
        else:
            my_data[i][4]=3
        #print (my_data[i])
        
        if (my_data[i][6]=='0'):
            x0.append(my_data[i])
        else:
            x1.append(my_data[i])
        
    ax0 = np.asarray(x0)
    ax1 = np.asarray(x1)
    a = len(ax0)
    b = len(ax1)
    #print (a)=500, b =810
    #print (my_data)
    
    np.random.shuffle( ax0 )
    train_x0 = ax0[ 0:375 , : ]
    test_x0 = ax0[ 375:500 , : ]
    np.random.shuffle( ax1 )
    train_x1 = ax1[ 0:608 , : ]
    test_x1 = ax1[ 608:810 , : ]
```

```python
# testcases= 202 + 125 = 327
# train = 608 + 375 = 983

#print (test_x1)
#print (test_x0)

train_data = np.concatenate((train_x0, train_x1), axis=0)
test_data = np.concatenate((test_x0, test_x1), axis=0)

print (train_data)
print (test_data)
```

```python
#with everything

for j in range(1,151,3):
    obj = knn(j)
    count=0
    for i in range(327):
        ans = obj.classification(train_data,test_data[i][:6])
        if ans != test_data[i][6]:
            count+=1
    print(count)
```
